Pivot table

A Pivot Table is an interactive way to quickly summarize large amounts of data.
A Pivot Table is used to summarise ,sort,reorganize,group,count,or total data stored in a table.

```python
import pandas as pd
df=pd.read_csv("Assignment2.csv")
df.pivot_table(index='Region', columns='Product',values='Price')
```

```python
df.pivot_table(index='Region',columns='Product',values='Price',aggfunc='sum')
```

```python
df.pivot_table(index='Region',columns='Product',values='Price',aggfunc='mean')
```

```python
df.pivot_table(index='Region', columns='Product', values='Price', aggfunc='max')
```

```python
df.pivot_table(index='Region', columns='Product',values='Price',aggfunc='min')
```
